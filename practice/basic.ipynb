# Preprocessing

```python
# preprocessing
# Remove stopwords

from nltk.corpus import stopwords
stop = stopwords.words('english')
data['tweet_text'] = data['tweet_text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))

# Remove mentions
data['tweet_text'] = data['tweet_text'].replace(r'@[A-Za-z0-9]+', '', regex=True)

# Remove URLs
data['tweet_text'] = data['tweet_text'].replace(r'http\S+', '', regex=True)

# Remove hashtags
data['tweet_text'] = data['tweet_text'].replace(r'#[A-Za-z0-9]+', '', regex=True)

# Remove special characters
data['tweet_text'] = data['tweet_text'].replace(r'[^A-Za-z0-9]+', ' ', regex=True)

# Remove numbers
data['tweet_text'] = data['tweet_text'].replace(r'\d+', '', regex=True)

# Remove extra spaces
data['tweet_text'] = data['tweet_text'].replace(r'\s+', ' ', regex=True)

# Convert to lowercase
data['tweet_text'] = data['tweet_text'].str.lower()

data
```

```python
import spacy
nlp = spacy.load('en_core_web_md')

# Lemmatization
data['tweet_text'] = data['tweet_text'].apply(lambda x: ' '.join([token.lemma_ for token in nlp(x)]))

data
```
