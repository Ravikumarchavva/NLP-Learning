```python
import pandas as pd

class Category:
    Books = 'books'
    Shoes = 'shoes'

data = pd.DataFrame({
    'text': [
        "I love books.", 
        "I hate shoes.", 
        "Books are great.", 
        "Shoes are uncomfortable.", 
        "Books bring joy.", 
        "Shoes make me tired.", 
        "I read books daily.", 
        "Shoes hurt my feet.", 
        "Reading books is fun.", 
        "Shoes are too tight.", 
        "Books are my passion.", 
        "Shoes are not my style.", 
        "Books help me learn.", 
        "Shoes are expensive.", 
        "I enjoy reading books.", 
        "I avoid wearing shoes.", 
        "Books give me peace.", 
        "Shoes are hard to clean.", 
        "Books inspire creativity.", 
        "Shoes are noisy.", 
    ],
    'sentiment': [
        Category.Books, 
        Category.Shoes, 
        Category.Books, 
        Category.Shoes, 
        Category.Books, 
        Category.Shoes, 
        Category.Books, 
        Category.Shoes, 
        Category.Books, 
        Category.Shoes, 
        Category.Books, 
        Category.Shoes, 
        Category.Books, 
        Category.Shoes, 
        Category.Books, 
        Category.Shoes, 
        Category.Books, 
        Category.Shoes, 
        Category.Books, 
        Category.Shoes, 
    ]
})

# Print the DataFrame
data
```

```python
import spacy
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression

# Load Spacy model
nlp = spacy.load('en_core_web_md')

# Tokenize the text and extract vectors
docs = list(nlp.pipe(data['text']))
X_vector = [doc.vector for doc in docs]

# Use string labels directly
y = data['sentiment']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_vector, y, test_size=0.2, random_state=42, shuffle=True)

# Train the RandomForest model
model = LogisticRegression()
model.fit(X_train, y_train)

# Predict the sentiment
y_pred = model.predict(X_test)

# Print the classification report
print(classification_report(y_test, y_pred))
```

```python
test_x = nlp("I love Sneakers.")
test_docs = list(nlp.pipe([test_x]))
test_X_vector = [doc.vector for doc in test_docs]

# Predict the sentiment
test_y_pred = model.predict(test_X_vector)
print(test_y_pred)

test_x = nlp("I love Library.")
test_docs = list(nlp.pipe([test_x]))
test_X_vector = [doc.vector for doc in test_docs]

# Predict the sentiment
test_y_pred = model.predict(test_X_vector)
print(test_y_pred)
```
